#9. Некоторые методы библиотеки  OpenCV

```python
%matplotlib inline
```

```python
import cv2
```

```python
from skimage import io
img = io.imread('1400022224_1683685347.jpg')
```

```python
io.imshow(img)
```

## Вот так можно выбирать кусок из изображения

```python
io.imshow(img[100:300,40:250])
```

##Так можно менять размер изображения

```python
miniImg = cv2.resize(img,(32,32))
```

```python
io.imshow(miniImg)
```

##Так можно из цветного изображения получать цветное

```python
gray_image = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)
```

```python
io.imshow(gray_image)
```

## Так можно рисовать на самом изображении ограничивающий прямоугольник

```python
cv2.rectangle(miniImg, (10,10), (20,20), (255,0,0))
```

```python
io.imshow(miniImg)
```
